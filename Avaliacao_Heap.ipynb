```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
import pandas as pd

import seaborn as sns

import math
sns.set()
```

```python
df_dados = pd.read_csv('dados.csv')
df_dados['Quantidade'] = df_dados['Quantidade'].apply(str)
#df_dados['Tempo'] = df_dados['Tempo'].transform(lambda x: truncate(x*1e2,3))
```

```python
df_dados.head(20)
```

```python
plt.figure(figsize=(20,10))
lineplot = sns.lineplot(data=df_dados, x='Quantidade', y='Tempo', hue='Estratégia', style="Estratégia", markers=True)
lineplot.get_figure().suptitle('Quantidade de itens X Tempo', fontsize=20)
_ = lineplot.set(xlabel='Quantidade', ylabel='Tempo (s)')
```

```python
fig = lineplot.get_figure()
fig.savefig("line_tempo.png")
```

```python
plt.figure(figsize=(20,10))
barplot = sns.barplot(data=df_dados, x='Quantidade', y='Tempo', hue='Estratégia')
_ = barplot.get_figure().suptitle('Quantidade de itens X Tempo', fontsize=20)
```

```python
fig = barplot.get_figure()
fig.savefig("bar_tempo.png")
```

```python
plt.figure(figsize=(20,10))
lineplot = sns.lineplot(data=df_dados, x='Quantidade', y='Comparações', hue='Estratégia', style="Estratégia", markers=True)
_ = lineplot.get_figure().suptitle('Quantidade de itens X Número de Comparações', fontsize=20)
```

```python
fig = lineplot.get_figure()
fig.savefig("line_comparacoes.png")
```
